## Build CSV

```python
import pandas as pd
import requests
import json
import os
```

```python
dt = pd.read_csv("../data/PASSON.csv",)
dt
```

```python
data_json = requests.get("https://172.16.13.224:8443/dms-api/public/v2/data?&from=2024-01-01T00:00:00Z&to=2024-04-01T00:00:00Z&sites=PASSON&dataTypes=base&groups=",
            verify=False).json()
data_json
```

```python
os.remouve('test_csv_append.csv')
for i in range(6):
    main_df = pd.DataFrame(columns=['date','id','value','unit','state','validated'])
    df = pd.DataFrame(data_json["data"][i]["sta"]["data"])
    df["id"] = data_json["data"][i]["id"]
    df["unit"] = data_json["data"][i]["sta"]['unit']
    dum = pd.concat([main_df, df], axis=1)
    # add concatenation de dataframes
    dum.to_csv('test_csv_append.csv',
              mode='a',
              header=(not os.path.exists('test_csv_append.csv'))
              )
```
